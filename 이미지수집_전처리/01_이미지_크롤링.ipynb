```python
#http://certifiedcar.kr/buy/sub_list.php?cd_gubun=1
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys
import time
import os
import urllib.request
```

# 차량 모델 크롤링

```python
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)


keyword = ['현대 싼타페TM', '기아 셀토스','기아 K7 YG','현대 포터2','현대 아반때 AD',
          '기아 모닝 JA','르노삼성 QM6', '현대 쏘나타 DNB','기아 카니발','기아 쏘렌토',
          '현대 베뉴', '현대 그랜저 IG', '기아 봉고3' ,'기아 K3 BD','쉐보레 스파크',
          '쌍용 렉스턴 스포츠','기아 K5 JF','현대 팰리 세이드','현대 그랜드 스타렉스',
          '쌍용 티볼리','현대 코나','기아 레이','현대 LF 쏘나타','기아 모하비','기아 스포티지 QL',
          '현대 투싼 TL', '쌍용 코란도','제네시스 G80','현대 쏘나타 DNB 하이브리드',
          '현대 그랜저 하이브리드 IG','제네시스 G70','기아 니로','기아 K7 하이브리드 YG',
          '르노삼성 SM6','제네시스 G90','현대 코나 일렉트릭','쉐보레 트랙스','르노삼성 QM3',
          '쌍용 G4 렉스턴','현대 코나 하이브리드']


for j in range(len(keyword)):
    createFolder('./'+keyword[j])
    driver = webdriver.Chrome("C:/chromedriver_win32/chromedriver.exe")
    driver.implicitly_wait(3)


    print(keyword[j], '검색')
    driver.get('https://www.google.co.kr/imghp?hl=ko')

    Keyword=driver.find_element_by_xpath('//*[@id="sbtc"]/div/div[2]/input')
    Keyword.send_keys(keyword[j])

    driver.find_element_by_xpath('//*[@id="sbtc"]/button').click()


    print(keyword[j]+' 스크롤 중 .............')
    elem =  driver.find_element_by_tag_name("body")
    for i in range(60):
        elem.send_keys(Keys.PAGE_DOWN)
        time.sleep(0.1)

    try:
        driver.find_element_by_xpath('//*[@id="islmp"]/div/div/div/div[1]/div[4]/div[2]/input').click()
        for i in range(60):
            elem.send_keys(Keys.PAGE_DOWN)
            time.sleep(0.1)
    except:
        pass


    links=[]
    images = driver.find_elements_by_css_selector("img.rg_i.Q4LuWd")
    for image in images:
        if image.get_attribute('src')!=None:
            links.append(image.get_attribute('src'))

    print(keyword[j]+' 찾은 이미지 개수:',len(links))
    time.sleep(2)


    for k,i in enumerate(links):
        url = i
        start = time.time()
        urllib.request.urlretrieve(url, "./"+keyword[j]+"/"+keyword[j]+"_"+str(k+1)+".jpg")
        print(str(k+1)+'/'+str(len(links))+' '+keyword[j]+' 다운로드 중....... Download time : '+str(time.time() - start)[:5]+' 초')
    print(keyword[j]+' ---다운로드 완료---')

    driver.close()
```
